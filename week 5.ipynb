Aim: Create a series of plots to analyze a given dataset.

Description:
This experiment involves loading a dataset and visualizing it using different types of plots such as line plots, bar plots, and histograms. The purpose is to understand trends, distributions, and patterns in the data effectively.

Procedure / Algorithm:

Import the required libraries (Matplotlib, Pandas, NumPy).

Load or create a sample dataset.

Generate different types of plots: Line plot, Bar plot, Histogram, Pie chart, etc.

Add titles and labels to each plot.

Display the plots using plt.show().

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset
x = np.arange(1, 6)
y = np.array([10, 20, 15, 30, 25])

# Line Plot
plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o')
plt.title("Line Plot of Sample Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Bar Plot
plt.figure(figsize=(6,4))
plt.bar(x, y, color='orange')
plt.title("Bar Plot of Sample Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

                                     Expiriment 5.2        

Aim: Generate a subplot layout with various plot types (scatter, line, bar).

Description:
This experiment demonstrates how to display multiple plots in a single figure using subplots. Different visualization types like scatter, line, and bar charts are plotted together for better comparison and analysis.

Procedure / Algorithm:

Import Matplotlib and NumPy.

Create data arrays.

Use plt.subplot() or plt.subplots() to define the layout.

Plot different chart types (scatter, line, bar) in each subplot.

Display all plots together.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = np.array([10, 20, 15, 30, 25])

plt.figure(figsize=(10,4))

# Line Plot
plt.subplot(1,3,1)
plt.plot(x, y, marker='o')
plt.title("Line Plot")

# Scatter Plot
plt.subplot(1,3,2)
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")

# Bar Plot
plt.subplot(1,3,3)
plt.bar(x, y, color='green')
plt.title("Bar Plot")

plt.tight_layout()
plt.show()
```

                                    Experiment 5.3          

Aim: Visualize time-series data and customize axis labels and date formats.

Description:
This experiment focuses on creating and visualizing time-series data using Matplotlib. It demonstrates how to format the x-axis with dates and how to make the visualization more readable by rotating and formatting date labels.

Procedure / Algorithm:

Import Matplotlib and Pandas.

Create or load time-series data.

Plot the data using plt.plot().

Format the x-axis with appropriate date formatting using DateFormatter.

Rotate and adjust labels for better readability.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create time series data
dates = pd.date_range('2023-01-01', periods=10)
values = np.random.randint(10, 50, size=10)

plt.figure(figsize=(8,4))
plt.plot(dates, values, marker='o')
plt.title("Time Series Plot")
plt.xlabel("Date")
plt.ylabel("Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Aim: Create a 3D plot.

Description:
This experiment introduces 3D plotting using Matplotlib's mplot3d toolkit. It shows how to plot points and lines in a 3D coordinate system, allowing visualization of relationships among three variables.

Procedure / Algorithm:

Import Matplotlib and NumPy.

Create 3D axes using fig.add_subplot(projection='3d').

Define X, Y, Z data points.

Plot the data using ax.plot3D().

Add labels and show the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x, y, z, 'blue')

ax.set_title("3D Line Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```
